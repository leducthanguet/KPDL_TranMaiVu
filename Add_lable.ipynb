```python
from unidecode import unidecode
def check_N(str1,str2):
    if str1.find(str2)!=-1:
        return 1
    return 0
def check_T(str1,str2):
    str1 = str1.lower()
    if str1.find(str2)!=-1:
        return 1
    return 0
def check_KD(str1,str2):
    str1 = str1.lower()
    str1 = unidecode(str1)
    if str1.find(str2)!=-1:
        return 1
    return 0
def check_final(str1):
    #100% xuất sắc
    if check_N(str1,"Xuất sắc")==1 or check_N(str1,"Tuyệt hảo")==1:
        return 1
    #100% tốt
     
    #100%TB
    elif check_N(str1,".  Dễ chịu")==1 or check_N(str1,".  Tàm tạm")==1 or check_N(str1,".  Chấp nhận được")==1:
        return 3
    elif check_KD(str1,".  binh thuong")==1:
        return 3
    #100% Kém
    
    
    #100% Rất kém
    
    # >90% Xuất sắc
    
    # >90% tốt 
    
    # >90% TB
    
    # >90% Kém
    
    # >90% Rất kém
    
    
f = open("C:\\Users\\leduc\\Downloads\\Phan_tich_quan_diem\\result1.txt", "a",encoding="utf-8")
file = open("C:\\Users\\leduc\\Downloads\\Phan_tich_quan_diem\\test.txt",encoding="utf-8")
for line in file:
#     line = line.lower()
    if check_final(line)==1:
        line = "__label__xuat_sac " + line
    elif check_final(line)==2:
        line = "__label__tot " + line
    elif check_final(line)==3:
        line = "__label__trung_binh " + line
    elif check_final(line)==4:
        line = "__label__kem " + line
    elif check_final(line)==5:
        line = "__label__rat_kem " + line
    else :
        line = "__label " + line
    f.write(line)
f.close()
    
    
        
    


```
